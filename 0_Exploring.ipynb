# Demo scrape some full text

```python
import requests
from bs4 import BeautifulSoup
apikey = open('apikey.txt', 'r').read()
db = 'pmc'
paper_id = 'PMC3877685'
base = 'http://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?'
s = '{:s}db={:s}&id={:s}&api_key={:s}'.format(base, db, paper_id, apikey)
out = requests.get(s)
page_soup = BeautifulSoup(out.content, 'lxml')
```
